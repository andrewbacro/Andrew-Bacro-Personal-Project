```python
import os
import pandas as pd
new_path = "\\Users\\bacro\\OneDrive\\Desktop\\PierrepontDS_2\\Datasets"
os.chdir(new_path)
college = pd.read_csv('IPEDS_data.csv', encoding = "Latin-1")
college.head(5)
```

```python
college_name = college["Name"] 
```

```python
reading_25 = college["SAT Critical Reading 25th percentile score"]
reading_75 = college["SAT Critical Reading 75th percentile score"]
math_25 = college["SAT Math 25th percentile score"]
math_75 = college["SAT Math 75th percentile score"]
sat_score_list = []
tuple_score_list = []
for i in range(len(reading_25)):
    sum = ((reading_25.get(i) + reading_75.get(i)) / 2) + ((math_25.get(i) + math_75.get(i)) / 2)
    sat_score_list.append(sum)
    tuple_score_list.append((sum, college_name[i]))
```

```python
tuple_score_list
```

```python
mean_sat = {}                       #This contains a dictionary of scores with the name being the key and mean sat score being the value (DNU)
college_sat = []                    #This contains the scores (DNU)
list = []                           #This contains a list of tuples with the element 0 of the tuple being the name of the college and element 1 being the mean sat score
counter = 0
for i in range(len(college_name)):
    reading = (college.loc[counter][26] + college.loc[counter][27]) / 2
    math = (college.loc[counter][28] + college.loc[counter][29]) / 2
    mean_sat[college_name[i]] = reading + math
    counter += 1
for name in mean_sat:
    tuple = (name, mean_sat[name])
    list.append(tuple)
    college_sat.append(mean_sat[name])
```
